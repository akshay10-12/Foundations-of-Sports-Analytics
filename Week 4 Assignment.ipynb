## Part 2 - Regression Analysis 1 - Test of Home Game Advantage

### 1. In this regression analysis, you will try to determine if playing at home gives teams any advantage in their performance.

Run the following regressions where the variable “score_diff” is the dependent variable.


### a). Reg1_1: Include a single variable “home” as the independent variable.

```python
reg1_1 = sm.ols(formula = 'score_diff ~ home', data = NFL_Game).fit() 
print(reg1_1.summary())
```

### b). Reg1_2: Include the “home,” “stadium_capacity,” “stadium_neutral” variables and the interaction between “home” and “stadium_neutral” as the independent variables.

```python
reg1_2 = sm.ols(formula = 'score_diff ~ home+stadium_capacity+stadium_neutral+home*stadium_neutral', data = NFL_Game).fit() 
print(reg1_2.summary())
```

## Part 3 - Regression Analysis 2 -- Impact of Outside Factors on Scores

### In this regression analysis, you will investigate if the final score earned by each team is affected by outside factors such as the size and condition of the stadium as well as weather conditions at the stadium.

Run the following regressions where the variable “score” is the dependent variable.

#### a). Reg2_1: Include “season” and “home” variables as independent variables."

```python
reg2_1 = sm.ols(formula = 'score ~ season + home', data = NFL_Game).fit() 
print(reg2_1.summary() )
```

#### b). Reg2_2: Include “season,” “home,” “weather_temperature,” “weather_wind_mph,” and “weather_humidity” as independent variables.

```python
reg2_2 = sm.ols(formula = 'score ~ season + home + weather_temperature + weather_wind_mph + weather_humidity', data = NFL_Game).fit()
```

```python
reg2_2.summary()
```

#### c). Reg2_3: Include “season,” “home,” “weather_temperature,” “weather_wind_mph,” “weather_humidity,” “stadium_capacity,” “stadium_age”, “stadium_type,” “stadium_neutral”, and the interaction between “home” and “stadium_neutral” as independent variables.     

```python
reg2_3 = sm.ols(formula = 'score ~ season + home + weather_temperature + weather_wind_mph + weather_humidity + stadium_capacity + stadium_age + stadium_type + stadium_neutral + home*stadium_neutral', data = NFL_Game).fit() 
reg2_3.summary() 
```
